```
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```
lines = open("./worldcities.csv","r").readlines()
print(len(lines))
```

```
# "city","city_ascii","lat","lng","country","iso2","iso3","admin_name","capital","population","id"

lats = []
lngs = []
pops = []
for l in lines[1:]:
    parts = l.split(",")
    if len(parts) != 11:
        #print("error on line")
        continue
    lats.append( float( parts[2].replace( '"', "") ) )
    lngs.append( float( parts[3].replace( '"', "") ) )
    tmp = parts[9].replace( '"', "")
    if len(tmp) == 0:
      val = 0
    else:
      val = float(tmp) / 1000000.0
    pops.append( val )

lats.reverse()
lngs.reverse()
pops.reverse()
```

```
plt.plot( pops )
```

```
plt.scatter( lngs, lats, s=pops, c=pops, cmap='jet')
plt.colorbar()
```
