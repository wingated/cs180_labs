```
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
```

```
lines = open("./worldcities.csv","r").readlines()
print(len(lines))
```

```
lats = []
lngs = []

for l in lines[1:]:
    parts = l.split(",")
    if len(parts) != 11:
        #print("error on line")
        continue
    lats.append( float( parts[2].replace( '"', "") ) )
    lngs.append( float( parts[3].replace( '"', "") ) )

plt.scatter( lngs, lats, s=0.1 )
```
